```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

# ***Resolvido Preço***

```python
url = "https://www.kalunga.com.br/busca/notebook/1"
html = requests.get(url).text
soup = BeautifulSoup(html, 'html5lib')
#nomes = []
precos = []
for div in soup.findAll('div', attrs={'class':'blocoproduto'}):
    #nome = div.find('h1', attrs={'class':'text-left'})
    preco = div.find('span', attrs={'class': 'blocoproduto__text blocoproduto__text--bold blocoproduto__price'})
    #nomes.append(nomes.text)
    if preco is not None:
        precos.append(preco.text)
    else:
        precos.append(None)
```

# ***Resolvendo Nome***

```python
url = "https://www.kalunga.com.br/busca/notebook/1"
html = requests.get(url).text
soup = BeautifulSoup(html, 'html5lib')
nomes = []
precos = []
for div in soup.findAll('div', attrs={'class':'blocoproduto'}):
    nome = div.find('h2', attrs={'class':'blocoproduto__title'})
    preco = div.find('span', attrs={'class': 'blocoproduto__text blocoproduto__text--bold blocoproduto__price'})
    #nomes.append(nome.text)
    
    if nome is not None:
        nomes.append(nome.text)
    else:
        nomes.append(None)
        
    if preco is not None:
        precos.append(preco.text)
    else:
        precos.append(None)
```

```python
nomes
```

```python
nomes = pd.DataFrame(nomes,columns=['nome'])
precos = pd.DataFrame(precos,columns=['preco'])
print(nomes.shape)
print(precos.shape)
```

```python
produtos = pd.concat([nomes,precos],axis=1)
print(produtos.shape)
produtos
```
