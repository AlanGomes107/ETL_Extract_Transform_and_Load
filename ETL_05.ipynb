# ***Tabela Bidirecional***

```python
pd.crosstab(dataset['Platform'], dataset['Genre'])
```

```python
table = pd.crosstab(dataset['Platform'], dataset['Genre'])
```

```python
table
```

# ***Construindo tabelas a partir da Tabela Direcional (Linha soma 100%)***

```python
table.divide(table.sum(axis=1), axis=0)
```

# ***Construindo tabelas a partir da Tabela Direcional (Coluna soma 100%)***

```python
table.divide(table.sum(axis=0), axis=1)
```

# ***Gráfico de barras (Percentual por categoria)***

```python
table.plot(kind="bar", stacked=True, figsize=(8,7))
```

# ***Gráfico de barras Empilhado (Percentual por categoria)***

```python
table_perc = table.divide(table.sum(axis=1), axis=0)
table_perc.plot(kind="bar", stacked=True, figsize=(8,7))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numCols = dataset.columns[dataset.dtypes == 'float64']
scaler.fit(dataset[numCols])
dataset_norm=pd.DataFrame(scaler.transform(dataset[numCols]),columns=numCols)
```
