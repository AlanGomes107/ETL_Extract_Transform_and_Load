```python
atores = []
for cast in filmes.cast:
# Nao podemos remover os caracteres
   cast = cast[1:-1]
# vamos ignorar os valores ausentes
   if cast:
      cast = str(cast).replace(", Jr", " Jr")
      atores.extend(str(cast).split(","))
atores = [str(ator).replace("'", "").replace('"', "").strip() for ator in atores]
while "and" in atores:atores.remove("and")
```

```python
atores
```

```python
c = Counter(atores)
c.keys()
c.most_common(10)
```

# ***Remover Stopwords***

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords.words('english')
#stopwords.words('portuguese')
```

```python
titulo=titulos[0]
```

```python
titulo
```

```python
[palavra for palavra in titulo if palavra not in stopwords.words('english')]
```

```python
def remove_stopwords(palavras):
    return [palavra for palavra in palavras if palavra not in stopwords.words('english')]
titulos_sem_stopwords = titulos.apply(lambda x : remove_stopwords(x))
```

```python
print(titulos_sem_stopwords)
```

```python
titulos2 = []
for titulo in titulos_sem_stopwords:
    titulos2.extend(titulo)
```

```python
c = Counter(titulos2)
c.keys()
pd.DataFrame(c.most_common(10), columns=["palavra", "frequencia"])
```
